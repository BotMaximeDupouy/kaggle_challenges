```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.preprocessing import image_dataset_from_directory
from PIL import Image
import os
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
from tensorflow.image import flip_left_right, random_flip_up_down,flip_up_down,rot90, random_flip_left_right
```

# Load data

```python
path_data = "../data/dog_breed_identification/train/"
path_test = "../data/dog_breed_identification/test/"

df_classes = pd.read_csv('../data/dog_breed_identification/labels.csv')

# je creer un dictionnaire avec clé = nom image, value = race
dict_classes = {}
for row in range (df_classes.shape[0]):
    dict_classes[df_classes['id'][row]] = df_classes['breed'][row]
```

```python
# je récupère les especes classées par ordre alphbétique
breed = df_classes['breed'].to_list()
breed = sorted(list(set(breed)))
len(breed)
```

```python
# fonction qui permet de transformer les images en array (64,64,3); renvoie un array + label des races
def load_data():
    imgs=[]
    labels = []
    img_name = list(df_classes['id'])
    for img in img_name:
        image_path = path_data +img+'.jpg'
        if os.path.exists(image_path):
            image = Image.open(image_path)
            image = image.resize((64, 64))
            imgs.append(np.array(image))
            label = dict_classes[img]
            labels.append(label)
    return imgs, labels
```

```python
imgs, labels = load_data()
```

```python
# même fonction qu'au dessus mais renvoie array + nom de l'image
def load_data_test():
    imgs=[]
    path_test = '../data/dog_breed_identification/test/'
    img_name = os.listdir(path_test)
    for img in img_name:
        image_path = path_test +img
        if os.path.exists(image_path):
            image = Image.open(image_path)
            image = image.resize((64, 64))
            imgs.append(np.array(image))
    return imgs, img_name
df_test, test_name = load_data_test()
```

# my first model

Un premier modele pour voir ce que ça donne

```python
from tensorflow.keras import Sequential, layers
from tensorflow.keras import optimizers
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.layers.experimental.preprocessing import Rescaling

def load_own_model():

    model = Sequential()
    model.add(Rescaling(1./255, input_shape=(64,64,3)))

    model.add(layers.Conv2D(16, kernel_size=8, activation='relu'))
    model.add(layers.MaxPooling2D(1))
    
    model.add(layers.Conv2D(32, kernel_size=4, activation="relu"))
    model.add(layers.MaxPooling2D(1))

    #model.add(layers.Conv2D(32, kernel_size=4, activation="relu"))
    
    model.add(layers.Flatten())
    model.add(layers.Dense(20, activation='relu'))
    model.add(layers.Dense(120, activation='softmax'))
    
    opt = optimizers.Adam()
    model.compile(loss='categorical_crossentropy',
                  optimizer=opt,
                  metrics=['accuracy'])
    
    return model

model_homemade = load_own_model()
model_homemade.summary()
```

```python
es = EarlyStopping(monitor='val_accuracy', mode='max', patience=5, verbose=1, restore_best_weights=True)

history = model_homemade.fit(X_train, y_train,
         batch_size=16, 
         epochs=100, 
         validation_data=(X_val, y_val),
         callbacks=[es])
```

```python
model_homemade.evaluate(X_test, y_test)
```

# transfer learning VGG

Au vu des faible résultat, je vais tester un transfer learning avec un VGG 16

```python
from tensorflow.keras.applications.vgg16 import VGG16

def load_model():
    # $CHALLENGIFY_BEGIN
    model = VGG16(weights="imagenet", include_top=False, input_shape=X_train[0].shape)
    # $CHALLENGIFY_END
    
    return model
```

```python
model = load_model()
model.summary()
```

```python
def set_nontrainable_layers(model):
    # Set the first layers to be untrainable
    model.trainable = False
    
    return model

model
model = set_nontrainable_layers(model)

model.summary()
```

```python
from tensorflow.keras import layers, models

def add_last_layers(model):
    '''Take a pre-trained model, set its parameters as non-trainables, and add additional trainable layers on top'''
    # $CHALLENGIFY_BEGIN
    base_model = set_nontrainable_layers(model)
    flatten_layer = layers.Flatten()
    dense_layer = layers.Dense(500, activation='relu')
    prediction_layer = layers.Dense(120, activation='softmax')
    
    
    model = models.Sequential([
        base_model,
        flatten_layer,
        dense_layer,
        prediction_layer
    ])
    # $CHALLENGIFY_END
    return model
```

```python
model = add_last_layers(model)
model.summary()
```

```python
from tensorflow.keras import optimizers

def build_model():
    # $CHALLENGIFY_BEGIN    
    model = load_model()
    model = add_last_layers(model)
    
    opt = optimizers.Adam(learning_rate=1e-4)
    model.compile(loss='categorical_crossentropy',
                  optimizer=opt,
                  metrics=['accuracy'])
    return model
    # $CHALLENGIFY_END
```

```python
model = build_model()
```

```python
from tensorflow.keras.applications.vgg16 import preprocess_input
X_train = preprocess_input(X_train) 
X_val = preprocess_input(X_val)
X_test = preprocess_input(X_test)
```

```python
from tensorflow.keras.callbacks import EarlyStopping

es = EarlyStopping(monitor='val_accuracy', mode='max', patience=5, verbose=1, restore_best_weights=True)

history = model.fit(X_train, y_train, 
                    validation_data=(X_val, y_val), 
                    epochs=50, 
                    batch_size=16, 
                    callbacks=[es])
```

```python
model.evaluate(X_test,y_test)
```

# Data augmentation

Résultats toujours médiocre -> tentative de data augmentation + modele VGG

## Data augmentation

```python

def augment_flip_l_r(image, label):
    image = flip_left_right(image)
    image = random_flip_up_down(image)
    return image, label
```

```python

def augment_flip_u_d(image, label):
    image = random_flip_left_right(image)
    image = flip_up_down(image)
    image = rot90(image)
    return image, label
```

```python
path_data = "../data/dog_breed_identification/train/"
path_test = "../data/dog_breed_identification/test/"

df_classes = pd.read_csv('../data/dog_breed_identification/labels.csv')

df_classes
dict_classes = {}
for row in range (df_classes.shape[0]):
    dict_classes[df_classes['id'][row]] = df_classes['breed'][row]
```

```python
def load_data_augmentated():
    imgs=[]
    labels = []
    img_name = list(df_classes['id'])
    for img in img_name:
        image_path = path_data +img+'.jpg'
        if os.path.exists(image_path):
            image = Image.open(image_path)
            image = image.resize((64, 64))
            imgs.append(np.array(image))
            label = dict_classes[img]
            labels.append(label)
            img_1, label_1 = augment_flip_l_r(image, label)
            img_2, label_2 = augment_flip_u_d(image, label)
            imgs.append(img_1)
            imgs.append(img_2)
            labels.append(label_1)
            labels.append(label_2)
            
    return imgs, labels
```

```python
imgs, labels = load_data_augmentated()
```

```python
X = np.array(imgs)
y=labels
le = LabelEncoder()
le.fit(y)
y_encoded = le.transform(y)
y_encoded_ohe = to_categorical(y_encoded)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded_ohe, test_size=0.3, random_state=0)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=0)
```

## model entrainé sur data augmenté

```python
model_2 = build_model()
from tensorflow.keras.applications.vgg16 import preprocess_input
X_train = preprocess_input(X_train) 
X_val = preprocess_input(X_val)
X_test = preprocess_input(X_test)
```

```python
es = EarlyStopping(monitor='val_accuracy', mode='max', patience=5, verbose=1, restore_best_weights=True)

history = model_2.fit(X_train, y_train, 
                    validation_data=(X_val, y_val), 
                    epochs=50, 
                    batch_size=32, 
                    callbacks=[es])
```
